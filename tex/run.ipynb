```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import sys
import yaml

from common import get_sim_results


params = '''
---
C1: 2
C2: [1.7, 1.8, 2.1, 2.2]
A: 0.5
T: 1.1
a: 0.9
b: 4.7
p1: 0.8
p2: 4

dt: 0.1
t: 0
x: 0.5
y: 0.5
N: 200
'''
    
def get_config():
  config = yaml.load(params)

  keys = []
  values = []
  for k, v in config.items():
    keys.append(k)
    values.append(v if isinstance(v, list) else [v])

  res = []
  for v in itertools.product(*values):
    res.append(dict(zip(keys, v)))
  return res



def main():
  markers = itertools.cycle('oDs*')
  configs = get_config()
  plt.figure(figsize=[20, 10])
  for params in configs:
    res = get_sim_results(**params)
    t = [v['t'] for v in res]
    x = [v['x'] for v in res]
    y = [v['y'] for v in res]
    plt.plot(t, x, marker=next(markers))
    plt.plot(t, y, linestyle='--')
  plt.ylabel('$x$')
  plt.xlabel('Time($t$)')
  plt.grid()
  plt.show()

main()
```
